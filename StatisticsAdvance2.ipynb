## Theory Questions and Answers

#### Q.1 What is hypothesis testing in statistics?
Ans. It is a claim or statement or an assumption about a population parameter that can be tested using statistical method.

for example: Average salary of an IT employee is 50K

#### Q.2  What is the null hypothesis, and how does it differ from the alternative hypothesis?
Ans. The null hypothesis is a statement that assumes there is no effect, no difference, or no relationship between variables. It represents the default or status quo assumption in hypothesis testing. The purpose of hypothesis testing is to determine whether there is enough statistical evidence to reject the null hypothesis.
The alternative hypothesis (H₁ or Ha) is the opposite of the null hypothesis. It suggests that there is an effect, a difference, or a relationship between variables. This is what researchers aim to support with data.

#### Q.3 What is the significance level in hypothesis testing, and why is it important?
Ans. The significance level (α) is the probability of rejecting the null hypothesis (H₀) when it is actually true. It helps decide whether to accept or reject H₀.
*Importance of Hypothesis Testing:-*
Controls Errors: Limits the chance of a Type I error (false positive).
Common Values: Usually 0.05 (5%) or 0.01 (1%).
Decision Rule: If p-value < α, reject H₀; otherwise, fail to reject H₀.

#### Q.4 What does a P-value represent in hypothesis testing?
Ans. The P-value is the probability of getting the observed results (or more extreme ones) if the null hypothesis (H₀) is true.
Importance:-
*Decision Rule:*
If P-value < significance level (α) → Reject H₀.
If P-value ≥ α → Fail to reject H₀.

*Indicates Strength of Evidence:*
A small P-value (e.g., < 0.05) suggests strong evidence against H₀.
A large P-value (e.g., > 0.05) means there is not enough evidence to reject H₀.

#### Q.5 How do you interpret the P-value in hypothesis testing?
Ans. The P-value tells us how likely we are to get the observed data if the null hypothesis (H₀) is true.

Interpretation:
1. P-value < α (e.g., 0.05) → Reject H₀
    Strong evidence against H₀.
    The alternative hypothesis (H₁) is supported.

2. P-value ≥ α → Fail to reject H₀
    Not enough evidence to reject H₀.
    Does not prove H₀ is true, just that we lack strong evidence against it.

#### Q.6  What are Type 1 and Type 2 errors in hypothesis testing?
Ans. *Type 1 Error (False Positive):*
Rejecting H₀ when it is actually true.
Example: A test says a drug works when it actually doesn’t.
Probability = Significance level (α) (e.g., 0.05 or 5%).

*Type 2 Error (False Negative):*
Failing to reject H₀ when it is actually false.
Example: A test says a drug doesn’t work when it actually does.
Probability = β (related to statistical power).

#### Q.7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
Ans. *One-Tailed Test:*
Tests for an effect in one specific direction (greater or smaller).
H₀: The new method is not better than the old method.
H₁: The new method is better than the old method.
Example: Checking if a new drug increases blood pressure (only in one direction).

*Two-Tailed Test:*
Tests for an effect in both directions (greater or smaller).
H₀: There is no difference between two methods.
H₁: There is a difference (could be higher or lower).
Example: Checking if a new drug affects blood pressure (either increase or decrease).

#### Q.8 What is the Z-test, and when is it used in hypothesis testing?
Ans. A Z-test is a statistical test used to determine whether two population means are different when the sample size is large (n ≥ 30) and the population variance is known.

*When to Use Z-Test?*
Large Sample Size (n ≥ 30).
Population Variance (σ²) is Known.
Data Follows a Normal Distribution.

#### Q.9  How do you calculate the Z-score, and what does it represent in hypothesis testing?

        z = (sample mean - population mean)/ population standard deviation

#### Q.10  What is the T-distribution, and when should it be used instead of the normal distribution?

*When to Use the T-Distribution?*
Small Sample Size (n < 30).

#### Q.12 What is the T-test, and how is it used in hypothesis testing?

#### Q.13 What is the relationship between Z-test and T-test in hypothesis testing?
Ans.Both Z-test and T-test are used to compare means and test hypotheses, but they differ based on sample size and whether the population standard deviation (σ) is known.

#### Q.14 What is a confidence interval, and how is it used to interpret statistical results?
Ans.A confidence interval (CI) is a range of values that estimates a population parameter (e.g., mean or proportion) with a certain level of confidence.

If a 95% CI for average height is (165 cm, 175 cm), it means:
We are 95% confident that the true population mean height falls between 165 cm and 175 cm.
A wider CI means more uncertainty; a narrower CI means more precision.

#### Q.15  What is the margin of error, and how does it affect the confidence interval?
Ans. The margin of error (MoE) is the amount of uncertainty in an estimate. It shows how much the sample result may differ from the true population value.
1. Larger MoE → Wider Confidence Interval → More uncertainty.
2. Smaller MoE → Narrower Confidence Interval → More precision.
3. Factors Affecting MoE:
    Higher confidence level (e.g., 99%) → Larger MoE (wider CI).
    Larger sample size (n) → Smaller MoE (narrower CI).
    Higher variability (σ) → Larger MoE.

#### Q.17 What is the Chi-square distribution, and when is it used?
Ans.The Chi-square (χ²) distribution is a probability distribution used to analyze categorical data and test relationships between variables. It is right-skewed and depends on degrees of freedom (df).

*When is Chi-Square Used?*
1. Chi-Square Goodness-of-Fit Test – Checks if a sample follows an expected distribution.
2. Chi-Square Test for Independence – Tests if two categorical variables are related.
3. Chi-Square Test for Variance – Compares a sample variance to a known population variance.

#### Q.18 What is the Chi-square goodness of fit test, and how is it applied?
Ans.The Chi-square goodness-of-fit test checks whether a sample follows a specified distribution. It compares the observed and expected frequencies to see if the difference is significant.

*How to Apply the Test?*
1. Define Hypotheses:
    H₀ (Null Hypothesis): The sample follows the expected distribution.
    H₁ (Alternative Hypothesis): The sample does not follow the expected distribution.
3. Calculate Expected Frequencies.
4. Compute the Chi-square Statistic (χ²).
5. Compare χ² with Critical Value or Use P-value.
    If P-value < 0.05, reject H₀ → The sample does not follow the expected distribution.

#### Q.19 What is the F-distribution, and when is it used in hypothesis testing?
Ans.The F-distribution is a probability distribution used primarily in comparing variances. It is always positively skewed and depends on two degrees of freedom (df₁, df₂).

*When is the F-Distribution Used?*
1. ANOVA (Analysis of Variance): Compares means of multiple groups by analyzing variances.
2. F-Test for Variance: Compares two population variances to check if they are equal.
3. Regression Analysis: Tests the overall significance of a regression model.

#### Q.20 What is an ANOVA test, and what are its assumptions?
Ans.ANOVA is a statistical test used to compare the means of three or more groups to determine if there is a significant difference between them. It uses the F-distribution to test the null hypothesis.

*Assumptions of ANOVA:*
1. Independence – Observations must be independent.
2. Normality – Data in each group should follow a normal distribution.

#### Q.21 What are the different types of ANOVA tests?
Ans. 
1. *One-Way ANOVA*
Compares means of three or more groups based on one independent variable.
Example: Comparing test scores of students from three different schools.

2. *Two-Way ANOVA*
Compares means based on two independent variables (factors).
Example: Studying the effect of study method and gender on test scores.

3. *Repeated Measures ANOVA*
Used when the same subjects are measured multiple times.
Example: Measuring blood pressure before, during, and after a treatment.

4. *MANOVA (Multivariate ANOVA)*
Compares means of multiple dependent variables at the same time.
Example: Studying how teaching style affects both math and science scores.

#### Q.22 What is the F-test, and how does it relate to hypothesis testing?
Ans.The F-test is a statistical test that compares variances of two populations to determine if they are significantly different. It follows the F-distribution and is commonly used in ANOVA and regression analysis.

*Hypotheses in an F-Test:*
1. H₀ (Null Hypothesis): The two variances are equal.
2. H₁ (Alternative Hypothesis): The variances are different.

```python
# Q.1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy.stats import norm

# Sample data
sample = [50, 52, 48, 47, 53, 49, 51, 50, 46, 52]  # Sample data points
population_mean = 50  # Known population mean
population_std = 4  # Known population standard deviation
n = len(sample)  # Sample size

# Calculate sample mean
sample_mean = np.mean(sample)

# Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Compute P-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05  # Significance level (5%)
if p_value < alpha:
    print("Result: Reject the null hypothesis (H₀). The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis (H₀). No significant difference between sample and population mean.")
```

```python
# Q.4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate a random sample
np.random.seed(42)  # For reproducibility
sample_size = 30
sample = np.random.normal(loc=55, scale=10, size=sample_size)  # Mean=55, Std=10

# Step 2: Define population parameters
population_mean = 50  # Known population mean
population_std = 10   # Known population standard deviation

# Step 3: Compute sample mean
sample_mean = np.mean(sample)

# Step 4: Compute Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Step 5: Compute P-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 6: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value: {p_value:.4f}")

# Step 7: Interpretation (Significance Level = 0.05)
alpha = 0.05
critical_value = norm.ppf(1 - alpha / 2)  # Two-tailed critical value

if abs(z_score) > critical_value:
    print("Result: Reject the null hypothesis (H₀). Significant difference.")
else:
    print("Result: Fail to reject the null hypothesis (H₀). No significant difference.")

# Step 8: Visualization of the decision region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)  # Standard normal distribution

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Normal Distribution", color="blue")
plt.axvline(x=-critical_value, color='red', linestyle='dashed', label="Critical Region (-1.96, 1.96)")
plt.axvline(x=critical_value, color='red', linestyle='dashed')
plt.axvline(x=z_score, color='green', linestyle='solid', label=f"Z-Score = {z_score:.2f}")

plt.fill_between(x, y, where=(x < -critical_value) | (x > critical_value), color='red', alpha=0.3)
plt.title("Two-Tailed Z-Test: Decision Region")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
# Q.5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_null=50, mu_alt=52, sigma=10, n=30, alpha=0.05):
    # Compute standard error
    se = sigma / np.sqrt(n)
    
    # Compute critical value for rejection region (for one-tailed test)
    critical_value = mu_null + norm.ppf(1 - alpha) * se
    
    # Compute Type 2 error (β) probability
    beta = norm.cdf(critical_value, loc=mu_alt, scale=se)

    # Generate x-values for plotting
    x = np.linspace(mu_null - 4 * se, mu_alt + 4 * se, 1000)
    
    # Compute normal distributions
    y_null = norm.pdf(x, mu_null, se)
    y_alt = norm.pdf(x, mu_alt, se)

    # Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_null, label="Null Hypothesis (H₀)", color="blue")
    plt.plot(x, y_alt, label="Alternative Hypothesis (H₁)", color="green")

    # Fill Type 1 error region (α - false positive)
    plt.fill_between(x, y_null, where=(x >= critical_value), color='red', alpha=0.3, label="Type 1 Error (α)")

    # Fill Type 2 error region (β - false negative)
    plt.fill_between(x, y_alt, where=(x < critical_value), color='orange', alpha=0.3, label="Type 2 Error (β)")

    # Mark critical value
    plt.axvline(critical_value, color="black", linestyle="dashed", label=f"Critical Value = {critical_value:.2f}")

    # Labels and legend
    plt.title("Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

    # Print Type 1 and Type 2 error probabilities
    print(f"Type 1 Error (α): {alpha:.3f}")
    print(f"Type 2 Error (β): {beta:.3f}")
    print(f"Power of the Test (1 - β): {1 - beta:.3f}")

# Call function with default values
visualize_type1_type2_errors()
```

```python
# Q.6 Write a Python program to perform an independent T-test and interpret the results
import numpy as np
from scipy.stats import ttest_ind

# Step 1: Generate two independent samples
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=55, scale=10, size=30)  # Sample 1: Mean=55, Std=10
group2 = np.random.normal(loc=50, scale=10, size=30)  # Sample 2: Mean=50, Std=10

# Step 2: Perform an independent T-test
t_stat, p_value = ttest_ind(group1, group2)

# Step 3: Print results
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Step 4: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (H₀). Significant difference between groups.")
else:
    print("Result: Fail to reject the null hypothesis (H₀). No significant difference between groups.")
```

```python
# Q.8 Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
import scipy.stats as stats

# Step 1: Generate random data for two independent samples
np.random.seed(42)  # For reproducibility
sample_size = 30
population_mean = 50
population_std = 10

# Generate samples
sample1 = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
sample2 = np.random.normal(loc=population_mean + 3, scale=population_std, size=sample_size)  # Mean shifted by 3

# Step 2: Perform a Two-Sample Z-test (Assuming population std is known)
z_score = (np.mean(sample1) - np.mean(sample2)) / (population_std * np.sqrt(2/sample_size))
p_value_z = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Step 3: Perform an Independent Two-Sample T-test (When population std is unknown)
t_stat, p_value_t = stats.ttest_ind(sample1, sample2)

# Step 4: Print Results
print(f"Z-Test Results:")
print(f"Z-Score: {z_score:.2f}")
print(f"P-Value (Z-Test): {p_value_z:.4f}")

print("\nT-Test Results:")
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value (T-Test): {p_value_t:.4f}")

# Step 5: Interpretation (Significance Level = 0.05)
alpha = 0.05
def interpret_test(p_value, test_name):
    if p_value < alpha:
        print(f"Result: Reject the null hypothesis (H₀). Significant difference detected by {test_name}.")
    else:
        print(f"Result: Fail to reject the null hypothesis (H₀). No significant difference detected by {test_name}.")

interpret_test(p_value_z, "Z-Test")
interpret_test(p_value_t, "T-Test")
```

```python
# Q.9  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    sample_mean = np.mean(sample)  # Calculate sample mean
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(sample)
    
    # Calculate the margin of error using t-distribution (for small samples)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    margin_of_error = t_critical * (sample_std / np.sqrt(sample_size))
    
    # Confidence Interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Generate random sample

ci = confidence_interval(sample_data)
print(f"95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
# Q.10 Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    sample_size = len(sample)
    
    # Get t-critical value (for small samples)
    t_critical = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)
    
    # Compute Margin of Error
    moe = t_critical * (sample_std / np.sqrt(sample_size))
    
    return moe

# Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Generate random sample

moe = margin_of_error(sample_data)
print(f"Margin of Error (95% Confidence Level): {moe:.2f}")
```

```python
# Q.11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A):
    # Total probability of evidence (law of total probability)
    prob_B = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * (1 - prior_A))
    
    # Apply Bayes' Theorem
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / prob_B

    return posterior_A_given_B

# Example: Medical Test for Disease
prior_disease = 0.01  # 1% of population has the disease
test_positive_given_disease = 0.95  # Sensitivity: 95% true positive rate
test_positive_given_no_disease = 0.05  # False positive rate: 5%

# Compute Posterior Probability
posterior_disease_given_positive = bayes_theorem(prior_disease, test_positive_given_disease, test_positive_given_no_disease)
print(f"Probability of having the disease given a positive test: {posterior_disease_given_positive:.4f}")
```

```python
# Q.12 Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import pandas as pd
import scipy.stats as stats

# Step 1: Create a Contingency Table (Observed Frequencies)
data = np.array([[30, 10],  # Category 1
                 [20, 40]]) # Category 2
table = pd.DataFrame(data, columns=["Group A", "Group B"], index=["Category X", "Category Y"])
print("Contingency Table:\n", table)

# Step 2: Perform the Chi-Square Test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)

# Step 3: Print Results
print(f"\nChi-Square Statistic: {chi2_stat:.2f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-Value: {p_value:.4f}")
print("\nExpected Frequencies:\n", pd.DataFrame(expected, columns=["Group A", "Group B"], index=["Category X", "Category Y"]))

# Step 4: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject the null hypothesis (H₀). The variables are dependent.")
else:
    print("\nResult: Fail to reject the null hypothesis (H₀). The variables are independent.")
```

```python
# Q.16  Implement an F-test using Python to compare the variances of two random samples

import numpy as np
import scipy.stats as stats

# Step 1: Generate Two Random Samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Size=30
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15, Size=30

# Step 2: Compute Sample Variances
var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased)
var2 = np.var(sample2, ddof=1)

# Step 3: Compute F-Statistic
F_stat = var1 / var2  # Larger variance should be in numerator for one-tailed test

# Step 4: Compute P-Value (Two-Tailed Test)
df1, df2 = len(sample1) - 1, len(sample2) - 1
p_value = 2 * min(stats.f.cdf(F_stat, df1, df2), 1 - stats.f.cdf(F_stat, df1, df2))

# Step 5: Print Results
print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-Statistic: {F_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 6: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject the null hypothesis (H₀). The variances are significantly different.")
else:
    print("\nResult: Fail to reject the null hypothesis (H₀). No significant difference in variances.")
```

```python
# Q.17  Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

import numpy as np
import scipy.stats as stats

# Step 1: Generate Random Sample Data (Three Groups)
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
group2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10
group3 = np.random.normal(loc=60, scale=10, size=30)  # Mean=60, Std=10

# Step 2: Perform One-Way ANOVA Test
F_stat, p_value = stats.f_oneway(group1, group2, group3)

# Step 3: Print Results
print(f"F-Statistic: {F_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 4: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject the null hypothesis (H₀). At least one group mean is significantly different.")
else:
    print("\nResult: Fail to reject the null hypothesis (H₀). No significant difference among group means.")
```

```python
# Q.19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

def check_anova_assumptions(groups):
    # Normality Check (Shapiro-Wilk Test)
    print("\n### Normality Check (Shapiro-Wilk Test) ###")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1}: W-Statistic = {stat:.4f}, P-Value = {p:.4f}")
        if p < 0.05:
            print(f"  -> Group {i+1} is NOT normally distributed ")
        else:
            print(f"  -> Group {i+1} is normally distributed ")

    # Homogeneity of Variance (Levene’s Test)
    print("\n### Homogeneity of Variances Check (Levene’s Test) ###")
    stat, p = stats.levene(*groups)
    print(f"Levene’s Statistic = {stat:.4f}, P-Value = {p:.4f}")
    if p < 0.05:
        print("  -> Variances are NOT equal (Assumption violated)")
```

```python
# Q.20  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Step 1: Generate Sample Data
np.random.seed(42)

# Factors: 'Diet' (A, B) and 'Exercise' (Low, High)
diet = np.repeat(['A', 'B'], 30)  # 2 levels, 30 samples each
exercise = np.tile(['Low', 'High'], 30)  # 2 levels, alternating

# Dependent Variable: Weight Loss (Simulated Data)
weight_loss = np.random.normal(5, 1.5, 30).tolist() + np.random.normal(7, 1.5, 30).tolist()

# Create DataFrame
df = pd.DataFrame({'Diet': diet, 'Exercise': exercise, 'Weight_Loss': weight_loss})

# Step 2: Perform Two-Way ANOVA
model = ols('Weight_Loss ~ C(Diet) + C(Exercise) + C(Diet):C(Exercise)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Step 3: Print ANOVA Table
print("### Two-Way ANOVA Results ###\n")
print(anova_table)

# Step 4: Interpretation
alpha = 0.05
for factor, p_val in anova_table['PR(>F)'].items():
    if p_val < alpha:
        print(f"\nFactor {factor} is **significant** (P-Value = {p_val:.4f}) ✅")
    else:
        print(f"\nFactor {factor} is **not significant** (P-Value = {p_val:.4f}) ❌")

# Step 5: Visualization - Boxplots
plt.figure(figsize=(10, 5))
sns.boxplot(x="Diet", y="Weight_Loss", hue="Exercise", data=df)
plt.title("Weight Loss by Diet & Exercise")
plt.xlabel("Diet Type")
plt.ylabel("Weight Loss")
plt.legend(title="Exercise Level")
plt.grid(True)
plt.show()

# Step 6: Interaction Plot
from statsmodels.graphics.factorplots import interaction_plot
plt.figure(figsize=(8, 5))
interaction_plot(df['Diet'], df['Exercise'], df['Weight_Loss'], colors=['red', 'blue'], markers=['D', 'o'], ms=10)
plt.title("Interaction Plot: Diet vs. Exercise on Weight Loss")
plt.show()
```

```python
# Q.21 Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Define F-distribution parameters (degrees of freedom)
df1 = 5  # Degrees of freedom for numerator
df2 = 20  # Degrees of freedom for denominator

# Step 2: Generate x values (F-statistic values)
x = np.linspace(0, 5, 1000)  # F-distribution typically starts from 0

# Step 3: Compute F-distribution PDF
y = stats.f.pdf(x, df1, df2)

# Step 4: Plot F-distribution
plt.figure(figsize=(8, 5))
plt.plot(x, y, color="blue", lw=2, label=f"F-distribution (df1={df1}, df2={df2})")
plt.fill_between(x, y, alpha=0.3, color="blue")  # Shaded area

# Step 5: Mark Critical Region (alpha = 0.05, one-tailed test)
alpha = 0.05
f_critical = stats.f.ppf(1 - alpha, df1, df2)
plt.axvline(x=f_critical, color="red", linestyle="dashed", label=f"Critical Value (α = {alpha})")

# Step 6: Labels & Legend
plt.title("F-Distribution & Critical Region")
plt.xlabel("F-Statistic")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)

# Step 7: Show Plot
plt.show()
```

```python
# Q.22 Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate Sample Data (3 Groups)
np.random.seed(42)
group_A = np.random.normal(50, 10, 30)  # Mean=50, Std=10, Size=30
group_B = np.random.normal(55, 10, 30)  # Mean=55, Std=10, Size=30
group_C = np.random.normal(60, 10, 30)  # Mean=60, Std=10, Size=30

# Step 2: Perform One-Way ANOVA
f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

# Step 3: Print Results
print("### One-Way ANOVA Results ###")
print(f"F-Statistic: {f_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 4: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (H₀). At least one group mean is different. ✅")
else:
    print("Result: Fail to reject the null hypothesis (H₀). No significant difference. ❌")

# Step 5: Create DataFrame for Visualization
df = pd.DataFrame({
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30,
    'Values': np.concatenate([group_A, group_B, group_C])
})

# Step 6: Boxplot Visualization
plt.figure(figsize=(8, 5))
sns.boxplot(x="Group", y="Values", data=df, palette="Set2")
plt.title("One-Way ANOVA: Group Mean Comparison")
plt.xlabel("Group")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
# Q.23 Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate Random Data from Normal Distribution
np.random.seed(42)
sample_1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Size=30
sample_2 = np.random.normal(loc=55, scale=10, size=30)  # Mean=55, Std=10, Size=30

# Step 2: Perform an Independent Two-Sample T-Test
t_stat, p_value = stats.ttest_ind(sample_1, sample_2)

# Step 3: Print Results
print("### Two-Sample T-Test Results ###")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 4: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (H₀). The means are significantly different. ✅")
else:
    print("Result: Fail to reject the null hypothesis (H₀). No significant difference. ❌")

# Step 5: Visualization - Distribution of Samples
plt.figure(figsize=(8, 5))
sns.histplot(sample_1, bins=10, kde=True, color='blue', label="Sample 1 (Mean=50)")
sns.histplot(sample_2, bins=10, kde=True, color='red', label="Sample 2 (Mean=55)")
plt.axvline(np.mean(sample_1), color='blue', linestyle='dashed', label=f"Mean 1 = {np.mean(sample_1):.2f}")
plt.axvline(np.mean(sample_2), color='red', linestyle='dashed', label=f"Mean 2 = {np.mean(sample_2):.2f}")
plt.title("Distribution of Two Samples")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q.24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
import scipy.stats as stats

# Step 1: Generate Sample Data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Size=30

# Step 2: Define Population Variance (Known)
pop_variance = 100  # Assumed population variance (Std^2)

# Step 3: Compute Sample Variance
sample_variance = np.var(sample, ddof=1)  # Sample variance with Bessel's correction
n = len(sample)  # Sample size

# Step 4: Compute Chi-Square Test Statistic
chi_square_stat = (n - 1) * sample_variance / pop_variance

# Step 5: Compute P-Value (Two-Tailed Test)
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df=n-1), 1 - stats.chi2.cdf(chi_square_stat, df=n-1))

# Step 6: Print Results
print("### Chi-Square Test for Population Variance ###")
print(f"Sample Variance: {sample_variance:.2f}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 7: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (H₀). The sample variance is significantly different. ✅")
else:
    print("Result: Fail to reject the null hypothesis (H₀). No significant difference. ❌")
```

```python
# Q.25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats

# Step 1: Define Data (Successes & Sample Sizes)
success_1 = 45  # Successful cases in group 1
size_1 = 100    # Total cases in group 1

success_2 = 30  # Successful cases in group 2
size_2 = 100    # Total cases in group 2

# Step 2: Compute Sample Proportions
p1 = success_1 / size_1
p2 = success_2 / size_2

# Step 3: Compute Pooled Proportion
p_pool = (success_1 + success_2) / (size_1 + size_2)

# Step 4: Compute Standard Error
se = np.sqrt(p_pool * (1 - p_pool) * (1/size_1 + 1/size_2))

# Step 5: Compute Z-Score
z_score = (p1 - p2) / se

# Step 6: Compute P-Value (Two-Tailed Test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 7: Print Results
print("### Two-Sample Z-Test for Proportions ###")
print(f"Proportion 1: {p1:.4f}")
print(f"Proportion 2: {p2:.4f}")
print(f"Z-Score: {z_score:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 8: Interpretation (Significance Level = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (H₀). The proportions are significantly different. ✅")
else:
    print("Result: Fail to reject the null hypothesis (H₀). No significant difference. ❌")
```
